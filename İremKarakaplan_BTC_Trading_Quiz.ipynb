```python
def calculate_indicators(df):
    df_copy = df.copy()

    period = 10
    multiplier = 3.0

    high_low = df_copy['High'] - df_copy['Low']
    high_close = (df_copy['High'] - df_copy['Close'].shift()).abs()
    low_close = (df_copy['Low'] - df_copy['Close'].shift()).abs()
    ranges = pd.concat([high_low, high_close, low_close], axis=1)
    true_range = ranges.max(axis=1)
    atr = true_range.ewm(alpha=1/period, adjust=False).mean()

    final_upper_band = (df_copy['High'] + df_copy['Low']) / 2 + (multiplier * atr)
    final_lower_band = (df_copy['High'] + df_copy['Low']) / 2 - (multiplier * atr)

    supertrend = pd.Series(index=df_copy.index, dtype=float)
    is_uptrend = pd.Series(True, index=df_copy.index)

    for i in range(1, len(df_copy)):
        if df_copy['Close'][i-1] > supertrend[i-1]:
            is_uptrend[i] = True
        elif df_copy['Close'][i-1] < supertrend[i-1]:
            is_uptrend[i] = False
        else:
            is_uptrend[i] = is_uptrend[i-1]

        if is_uptrend[i]:
            supertrend[i] = final_lower_band[i]
        else:
            supertrend[i] = final_upper_band[i]
            
    df_copy['Supertrend'] = supertrend
    df_copy['Supertrend_direction'] = is_uptrend.apply(lambda x: 1 if x else -1)

    df_copy['KAMA'] = ta.momentum.KAMAIndicator(close=df_copy['Close'], window=10, pow1=2, pow2=30).kama()

    bb = ta.volatility.BollingerBands(close=df_copy['Close'], window=20, window_dev=2)
    df_copy['BB_lower'] = bb.bollinger_lband()
    df_copy['BB_upper'] = bb.bollinger_hband()
    df_copy['BB_middle'] = bb.bollinger_mavg()
    
    try:
        df_copy['RSI'] = ta.momentum.rsi(close=df_copy['Close'], window=14)
    except AttributeError:
        df_copy['RSI'] = ta.momentum.RSIIndicator(close=df_copy['Close'], window=14).rsi()

    macd = ta.trend.MACD(close=df_copy['Close'], window_fast=12, window_slow=26, window_sign=9)
    df_copy['MACD'] = macd.macd()
    df_copy['MACD_signal'] = macd.macd_signal()
    df_copy['MACD_diff'] = macd.macd_diff()

    df_copy = df_copy.ffill()
    df_copy = df_copy.dropna()
    
    return df_copy

price_df_with_indicators = calculate_indicators(price_df)

print("\n--- Success (Enhanced Manual SuperTrend) ---")
if price_df_with_indicators.empty:
    print("Error: DataFrame is empty")
else:
    print(price_df_with_indicators.tail())
```

```python
def determine_entry_exit_timing(event_type):
    
    if event_type in ['CPI', 'Fed_Rate', 'Nonfarm_Payrolls', 'Unemployment']:
        days_before_entry = 2
        days_after_exit = 2
    elif event_type == 'Jobless_Claims':
        days_before_entry = 1
        days_after_exit = 1
    else:
        days_before_entry = 1
        days_after_exit = 1
            
    return days_before_entry, days_after_exit


def generate_signal(row):

    rsi_oversold = row['RSI'] < 44
    price_above_supertrend = row['Close'] > row['Supertrend']
    
    if rsi_oversold and price_above_supertrend:
        signal = "BUY"
        indicator_used = "RSI+Supertrend"
        indicator_value = f"RSI:{row['RSI']:.2f}, Close:{row['Close']:.2f}, Supertrend:{row['Supertrend']:.2f}"
        rule_condition = (f"RSI={row['RSI']:.2f} < 44 (Oversold) AND "
                          f"Close > Supertrend (Uptrend Confirmed)")
    else:
        signal = "HOLD"
        indicator_used = None
        indicator_value = None
        rule_condition = "No signal triggered"

    return signal, indicator_used, indicator_value, rule_condition
```

```python
def backtest_strategy(price_data, events_data, initial_balance=10000):

    trade_log = []
    cumulative_balance = float(initial_balance)
    trade_no = 0

    for index, event in events_data.iterrows():
        event_date = event['date']
        event_type = event['Event_Type']

        days_before_entry, days_after_exit = determine_entry_exit_timing(event_type)

        potential_entry_date = event_date - pd.Timedelta(days=days_before_entry)
        
        if potential_entry_date < price_data.index.min():
            print(f"Error for date: {potential_entry_date.date()}")
            continue
        entry_date = price_data.index.asof(potential_entry_date)

        entry_row = price_data.loc[entry_date]
        signal, indicator_used, indicator_value, rule_condition = generate_signal(entry_row)

        if signal == "BUY":
            trade_no += 1
            entry_price = entry_row['Close']
            
            position_size_usdt = cumulative_balance
            position_size_btc = position_size_usdt / entry_price

            potential_exit_date = event_date + pd.Timedelta(days=days_after_exit)

            if potential_exit_date > price_data.index.max():
                exit_date = price_data.index.max()
                print(f"Warning for the date: {potential_exit_date.date()}. Last available date {exit_date.date()} is using.")
            else:
                exit_date = price_data.index.asof(potential_exit_date)

            exit_price = price_data.loc[exit_date]['Close']

            profit_loss_usdt = (exit_price - entry_price) * position_size_btc
            profit_loss_pct = (profit_loss_usdt / position_size_usdt) * 100
            
            cumulative_balance += profit_loss_usdt

            trade_log.append({
                'Trade_No': trade_no,
                'Event_Type': event_type,
                'Event_Date': event_date.strftime('%Y-%m-%d'),
                'Days_Before_Entry': days_before_entry,
                'Entry_Date': entry_date.strftime('%Y-%m-%d'),
                'Entry_Price': entry_price,
                'Days_After_Exit': days_after_exit,
                'Exit_Date': exit_date.strftime('%Y-%m-%d'),
                'Exit_Price': exit_price,
                'Indicator_Used': indicator_used,
                'Indicator_Value': indicator_value,
                'Rule_Condition': rule_condition,
                'Signal': signal,
                'Position_Size_USDT': position_size_usdt,
                'Position_Size_BTC': position_size_btc,
                'Profit_Loss_USDT': profit_loss_usdt,
                'Profit_Loss_Pct': profit_loss_pct,
                'Cumulative_Balance': cumulative_balance
            })
        
        else:
            pass

    return pd.DataFrame(trade_log)

trade_results_df = backtest_strategy(price_df_with_indicators, events_df, initial_balance=10000)

print("\n--- Backtesting Results ---")
print(f"Total {len(trade_results_df)} stock market transaction")
display(trade_results_df)
```

```python
def generate_performance_report(trades_df, initial_capital=10000):
    if trades_df.empty:
        print("No trades found to analyze.")
        return

    total_trades = len(trades_df)
    winning_trades = trades_df[trades_df['Profit_Loss_USDT'] > 0]
    losing_trades = trades_df[trades_df['Profit_Loss_USDT'] <= 0]
    
    num_winning_trades = len(winning_trades)
    num_losing_trades = len(losing_trades)
    
    win_rate = (num_winning_trades / total_trades) * 100 if total_trades > 0 else 0
    
    total_profit_loss = trades_df['Profit_Loss_USDT'].sum()
    final_balance = initial_capital + total_profit_loss
    roi = (total_profit_loss / initial_capital) * 100

    avg_pl_per_trade = trades_df['Profit_Loss_USDT'].mean()
    best_trade = trades_df['Profit_Loss_USDT'].max()
    worst_trade = trades_df['Profit_Loss_USDT'].min()
    
    avg_entry_timing = trades_df['Days_Before_Entry'].mean()
    avg_exit_timing = trades_df['Days_After_Exit'].mean()

    most_profitable_indicator = trades_df.groupby('Indicator_Used')['Profit_Loss_USDT'].sum().idxmax()

    print("\n--- STRATEGY PERFORMANCE REPORT ---")
    print("\n--- Basic Metrics ---")
    print(f"Total Number of Trades Executed: {total_trades}")
    print(f"Number of Winning Trades: {num_winning_trades}")
    print(f"Number of Losing Trades: {num_losing_trades}")
    print(f"Win Rate: {win_rate:.2f}%")
    print(f"Total Profit/Loss: ${total_profit_loss:,.2f} USDT")
    print(f"Starting Capital: ${initial_capital:,.2f} USDT")
    print(f"Final Balance: ${final_balance:,.2f} USDT")
    print(f"Return on Investment (ROI): {roi:.2f}%")
    
    print("\n--- Advanced Metrics ---")
    print(f"Average P/L per Trade: ${avg_pl_per_trade:,.2f} USDT")
    print(f"Best Trade (Highest Profit): ${best_trade:,.2f} USDT")
    print(f"Worst Trade (Largest Loss): ${worst_trade:,.2f} USDT")
    print(f"Average Entry Timing (Days Before Event): {avg_entry_timing:.2f}")
    print(f"Average Exit Timing (Days After Event): {avg_exit_timing:.2f}")
    print(f"Most Profitable Indicator Combination: {most_profitable_indicator}")
    print("-------------------------------------")

generate_performance_report(trade_results_df, initial_capital=10000)
```

```python

Strategy Documentation & Rationale
This notebook implements an event-driven trading strategy for the BTC/USDT pair, designed to capitalize on market volatility surrounding major U.S. macroeconomic data releases.

1. Entry & Exit Timing Strategy
I have chosen an Event-Type Strategy for both entering and exiting trades.

Rationale: High-impact events (CPI, Fed Rate) often cause more significant market reactions than medium-impact events (Jobless Claims). This strategy adapts to the expected impact of each event.
Implementation:
High-Impact Events (CPI, Fed Rate, NFP, Unemployment): Enter 2 days before and exit 2 days after.
Medium-Impact Events (Jobless Claims): Enter 1 day before and exit 1 day after.
2. Signal Generation Strategy & Parameter Optimization
For generating buy signals, I selected the Multiple Indicator Confirmation approach.

Rationale: Requiring confirmation from two different types of indicators—one for momentum (RSI) and one for trend (Supertrend)—increases the confidence and quality of each trade entry.
Implementation (Trading Rule):
A "BUY" signal is generated only when BOTH of the following conditions are met:
RSI < 44: The Relative Strength Index must be below 44.
Close Price > Supertrend: The closing price must be above the Supertrend line, confirming an overall bullish trend.
Parameter Optimization to Meet Minimum Trade Requirement:
The initial backtest using the standard RSI oversold threshold of < 30 yielded only a few trades, failing to meet the project's minimum requirement of 10 executed trades.
Following the suggestion in the project FAQ to adjust indicator thresholds, I systematically increased the RSI value. After testing thresholds of 35 and 40, which still produced fewer than 10 trades, I found that an RSI threshold of < 44 successfully generated a sufficient number of trades. This adjustment allows the strategy to be more sensitive to buying opportunities while still adhering to its core logic of combining momentum and trend confirmation. I have documented this change as a necessary step to fulfill the assignment's core requirements.
3. Expected Outcome
My objective with this adjusted strategy is to meet the minimum trade requirement while still maintaining a logical, rules-based approach. By relaxing the entry condition to RSI < 44, I expect to see more frequent trading activity, which provides a richer dataset for a comprehensive performance analysis.
```
